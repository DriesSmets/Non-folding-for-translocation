# Refolding with Circular Dichroism
CD spectra were recorded at 222 nm (minima for PpiA and PpiB). Denatured proteins (6M Urea) were diluted in aqueous buffer to initiate folding at 4Â°C (0.2M Urea).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
```

```python
spectra = np.genfromtxt('CD_PpiB_folding_4C.txt').T
spectra.shape
```

```python
#plot CD spectra of the two datasets
plt.xlabel('Time(sec)')
plt.ylabel('mdeg')
plt.plot(*spectra)
```

```python
#plot data first sample
x1=spectra[0]
y1=spectra[1]
# Design the Buterworth filter
N  = 3    # Filter order
Wn = 0.005 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data1 = signal.filtfilt(B,A,y1)

fig, (ax) = plt.subplots()
ax.plot(x1,y1)
ax.plot(x1,smooth_data1)
ax.set_xlabel('Time(min)')
ax.set_ylabel('mdeg')

plt.show()
```

```python
np.savetxt('PpiB folding 1', smooth_data1)
```

```python
#plot data second sample
x2=spectra[2]
y2=spectra[3]
# Design the Buterworth filter
N  = 3    # Filter order
Wn = 0.005 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data2 = signal.filtfilt(B,A,y2)


fig,ax1=plt.subplots()
ax1.plot(x2,y2)
ax1.plot(x2,smooth_data2)
plt.xlabel('Time(min)')
plt.ylabel('mdeg')
plt.show()
```

```python
np.savetxt('PpiB folding 2', smooth_data2)
```
