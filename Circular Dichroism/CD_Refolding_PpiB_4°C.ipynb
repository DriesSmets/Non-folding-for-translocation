# Refolding with Circular Dichroism (222 nm)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
```

```python
spectra = np.genfromtxt('PpiB folding.txt').T
spectra.shape
```

```python
#plot raw data of PpiB refolding at 4Â°C
x1=spectra[0]
y1=spectra[1]
plt.xlabel('Time(min)')
plt.ylabel('delta ellipticity')
plt.plot (x1,y1)
```

```python
# Fitting raw data with smoothed curve from Butterworth filter
N  = 3    # Filter order
Wn = 0.005 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data1 = signal.filtfilt(B,A,y1)


fig,ax1=plt.subplots()
ax1.plot(x1,y1)
ax1.plot(x1,smooth_data1)
plt.xlabel('Time(min)')
plt.ylabel('delta ellipticity')
plt.show()
```
