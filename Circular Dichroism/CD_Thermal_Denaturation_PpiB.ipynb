```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import peakutils
from scipy.signal import savgol_filter
import scipy.signal as signal
import pandas as pd
import statistics 
```

```python
# update to ppia
spectra = np.genfromtxt('Thermal_PpiB_208.txt').T
spectra.shape
```

```python
#plot all raw data
x=spectra[0]
y=spectra[1]
plt.xlabel('Temperature')
plt.ylabel('mmdeg')
plt.plot(*spectra)
```

```python
#plot data first sample
x1=spectra[0]
y1=spectra[1]
# Design the Buterworth filter
N  = 3    # Filter order
Wn = 0.1 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data1 = signal.filtfilt(B,A,y1)

yprime1 = np.diff(smooth_data1)
xprime1 = x[:-1]+0.5*np.diff(x1)
fig,(ax1,ax2)=plt.subplots(2)
ax1.plot(x1,y1)
ax1.plot(x1,smooth_data1)
ax2.plot(xprime1,yprime1)
#plt.plot(x,y)
plt.xlabel('Temperature')
plt.ylabel('mmdeg')

plt.show()
np.argmax(yprime1)
xprime1[np.argmax(yprime1)]
```

```python
#plot data second sample
x2=spectra[2]
y2=spectra[3]
# Design the Buterworth filter
N  = 3    # Filter order
Wn = 0.1 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data2 = signal.filtfilt(B,A,y2)

yprime2 = np.diff(smooth_data2)
xprime2 = x[:-1]+0.5*np.diff(x2)
fig,(ax1,ax2)=plt.subplots(2)
#ax1.plot(x2,y2)
ax1.plot(x2,smooth_data2)
ax2.plot(xprime2,yprime2)
#plt.plot(x,y)
plt.xlabel('Temperature')
plt.ylabel('mdeg')

plt.show()
np.argmax(yprime2)
xprime2[np.argmax(yprime2)]
```

```python
#plot data third sample
x3=spectra[4]
y3=spectra[5]
# Design the Buterworth filter
N  = 3    # Filter order
Wn = 0.1 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data3 = signal.filtfilt(B,A,y3)

yprime3 = np.diff(smooth_data3)
xprime3 = x[:-1]+0.5*np.diff(x)
fig,(ax1,ax2)=plt.subplots(2)
ax1.plot(x3,y3)
ax1.plot(x3,smooth_data3)
ax2.plot(xprime3,yprime3)
#plt.plot(x,y)
plt.xlabel('Temperature')
plt.ylabel('mmdeg')

plt.show()
np.argmax(yprime3)
xprime3[np.argmax(yprime3)]
```
