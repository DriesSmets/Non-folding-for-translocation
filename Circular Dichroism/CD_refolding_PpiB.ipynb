```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
```

```python
spectra = np.genfromtxt('PpiB folding.txt').T
spectra.shape
```

```python
#plot all raw data
plt.xlabel('Time(sec)')
plt.ylabel('mdeg')
plt.plot(*spectra)
```

```python
# Derivatives are not needed here right?

#plot data first sample
x1=spectra[0]
y1=spectra[1]
# Design the Buterworth filter
N  = 3    # Filter order
Wn = 0.005 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data1 = signal.filtfilt(B,A,y1)

yprime1 = np.diff(smooth_data1)
xprime1 = x1[:-1]+0.5*np.diff(x1)
fig, (ax1,ax2) = plt.subplots(2)
ax1.plot(x1,y1)
ax1.plot(x1,smooth_data1)
ax2.plot(xprime1,yprime1)
#plt.plot(x,y)
ax1.set_xlabel('Time(min)')
ax1.set_ylabel('mdeg')

plt.show()
np.argmax(yprime1)
xprime1[np.argmax(yprime1)]
```

```python
np.savetxt('PpiB folding 1', smooth_data1)
```

```python
#plot data second sample
x2=spectra[2]
y2=spectra[3]
# Design the Buterworth filter
N  = 3    # Filter order
Wn = 0.005 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data2 = signal.filtfilt(B,A,y2)


fig,ax1=plt.subplots()
ax1.plot(x2,y2)
ax1.plot(x2,smooth_data2)
plt.xlabel('Time(min)')
plt.ylabel('mdeg')
plt.show()
```

```python
np.savetxt('PpiB folding 2', smooth_data2)
```
